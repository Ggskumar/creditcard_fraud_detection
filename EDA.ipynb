```python
X=df.drop('default payment next month',axis=1)
y=df['default payment next month']
```

```python
plt.figure(figsize=(15,30))
plotnumber=1
for i in X.columns:
    ax=plt.subplot(8,3,plotnumber)
    sns.histplot(df[i])
    plt.xlabel(i,)
    plotnumber+=1
plt.tight_layout()
plt.show()
```

1.On time payments are high when compared to delayed playements

2.Balance amount are left skewed for 6 months

3.payment amount for all 6 months are left skewed
